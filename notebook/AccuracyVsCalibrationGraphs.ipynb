<h1>Accuracy vs Calibration Graphs</h1>



```python
import matplotlib.pyplot as plt
import json

# Load JSON files
with open('../betting_simulation_results/betting_simulation_results_standard.json') as f:
    baseline = json.load(f)
with open('../betting_simulation_results/betting_simulation_results_deepseek.json') as f:
    deepseek = json.load(f)
with open('../betting_simulation_results/betting_simulation_results_openai.json') as f:
    gpt4o = json.load(f)
with open('../betting_simulation_results/betting_simulation_results_qwen.json') as f:
    qwen = json.load(f)
with open('../betting_simulation_results/betting_simulation_results_openai4.1.json') as f:
    gpt41 = json.load(f)


# Define configurations
betting_types = ["kelly_f1"]
models = ["Random Forest", "SVC", "Logistic Regression", "XGBoost"]
llms = {
    "Baseline": baseline,
    "DeepSeek": deepseek,
    "OpenAI GPT-4o mini": gpt4o,
    "OpenAI GPT-4.1": gpt41,
    "Qwen": qwen,
    
}

# Generate plots using loops
for betting_type in betting_types:
    for model in models:
        plt.figure(figsize=(12, 6))
        for llm_name, data in llms.items():
            plt.plot(
                data[betting_type][model], 
                label=f"{llm_name} {model}"
            )
        plt.title(
            f"Betting Simulation with {model} Model"
           
        )
        plt.xlabel("Match Number")
        plt.ylabel("Bankroll")
        plt.legend()
        plt.show()
```
