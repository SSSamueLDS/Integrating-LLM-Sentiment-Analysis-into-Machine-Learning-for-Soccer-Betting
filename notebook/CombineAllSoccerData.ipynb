Combining All - cumulative all, cumulative 5, odds, and winner

```python
import json
import datetime
mapping_dict = {
    'Arsenal' : 'Arsenal',
    'Aston Villa' : 'Aston Villa', 
    'Bournemouth':'Bournemouth', 
    'Brentford':'Brentford', 
    'Brighton':'Brighton', 
    'Burnley':'Burnley', 
    'Cardiff':'Cardiff', 
    'Chelsea':'Chelsea', 
    'Crystal Palace':'Crystal Palace', 
    'Everton':'Everton', 
    'Fulham':'Fulham', 
    'Huddersfield':'Huddersfield', 
    'Hull':'Hull City', 
    'Leeds':'Leeds', 
    'Leicester':'Leicester', 
    'Liverpool':'Liverpool', 
    'Luton':'Luton', 
    'Man City':'Manchester City', 
    'Man United':'Manchester United', 
    'Middlesbrough':'Middlesbrough', 
    'Newcastle':'Newcastle', 
    'Norwich':'Norwich', 
    "Nott'm Forest":'Nottingham Forest', 
    'Sheffield United':'Sheffield Utd', 
    'Southampton':'Southampton', 
    'Stoke':'Stoke City', 
    'Sunderland':'Sunderland', 
    'Swansea':'Swansea', 
    'Tottenham':'Tottenham', 
    'Watford':'Watford', 
    'West Brom':'West Brom', 
    'West Ham':'West Ham', 
    'Wolves':'Wolves'
}
for i in range(2015,2024):

    #Get all 3 files
    data1 = None
    data2 = None
    data3 = None
    data4 = None

    to_json = []
    with open(f'../dataset/epl_cumulative_diff/epl_cumulative_diff_{i}.json') as f:
        data1 = json.load(f)
    with open(f'../dataset/epl_cumulative_diff_5/epl_cumulative_diff_5_{i}.json') as f:
        data2 = json.load(f)
    with open(f'../dataset/epl_odds/epl_odds_{i}.json') as f:
        data3 = json.load(f)
    with open(f'../dataset/fixtures_epl/fixtures_epl_{i}.json') as f:
        data4 = json.load(f)
    print(len(data1),len(data2),len(data3),len(data4))
    #Combine data1 and data2 together first
    combined=[]
    
    for fixture in data2:
        matched_fixture = None
        for other_fixture in data1:
            if fixture['fixture_id'] == other_fixture['fixture_id']:
                matched_fixture = other_fixture
                break
        combined_fixture = {x:fixture[x] for x in fixture}

        for k in matched_fixture:
            if k not in combined_fixture:
                combined_fixture[k] = matched_fixture[k]

        #Get the win-loss result
        for third_fixture in data4:
            if fixture['fixture_id'] == third_fixture['fixture']['id']:
                combined_fixture['actual_home_goals'] = third_fixture['goals']['home']
                combined_fixture['actual_away_goals'] = third_fixture['goals']['away']
                if third_fixture['goals']['home'] > third_fixture['goals']['away']:
                    combined_fixture['actual_winner'] = 'home'
                elif third_fixture['goals']['home'] < third_fixture['goals']['away']:
                    combined_fixture['actual_winner'] = 'away'
                else:
                    combined_fixture['actual_winner'] = 'draw'
                break
    
        combined.append(combined_fixture)
    
    for fixture in combined:
        for other_fixture in data3:
            
            # compare timestamp(example 1442671200) vs a date string (example "08/08/2015" or "08/08/15") and see if they are on the same day
            import datetime

            def is_same_date(timestamp, date_str):
                try:
                    # Convert timestamp to UTC datetime
                    utc_date = datetime.datetime.utcfromtimestamp(timestamp)
                except ValueError:
                    # Handle invalid timestamp (unlikely for valid timestamps)
                    return False
                
                # Define possible date formats
                formats = ["%d/%m/%Y", "%d/%m/%y"]
                parsed_date = None
                
                for fmt in formats:
                    try:
                        parsed_date = datetime.datetime.strptime(date_str, fmt)
                        break
                    except ValueError:
                        continue
                
                if parsed_date is None:
                    # None of the formats matched
                    return False
                
                # Compare the date parts (year, month, day)
                return utc_date.date() == parsed_date.date()
            
            if is_same_date(fixture["timestamp"], other_fixture["Date"]) and fixture["home_team"] == mapping_dict[other_fixture["HomeTeam"]] and fixture["away_team"] == mapping_dict[other_fixture["AwayTeam"]]:

                combined_fixture = {x:fixture[x] for x in fixture}
                combined_fixture['home_odds'] = other_fixture['B365H']
                combined_fixture['draw_odds'] = other_fixture['B365D']
                combined_fixture['away_odds'] = other_fixture['B365A']


                to_json.append(combined_fixture) 

                break
    # Create file
    #save the file
    print(len(to_json))
    json_string = json.dumps(to_json, indent=4)
    
    with open(f'epl_combined/epl_combined_{i}.json','w') as outfile:
        outfile.write(json_string)
        print(str(i)+'success')


            


            
```
